```python
from tensorflow.keras.initializers import glorot_uniform
```

```python
def getModel(dropout=0.1, neurons_first=500, neurons_second=250, learningRate=0.1):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Input((21 * 3 * 100 + 3, )),)
    model.add(tf.keras.layers.Dense(neurons_first, activation='relu', input_dim=6304, 
                    kernel_initializer=glorot_uniform(),
                    name='Dense_first'))
    model.add(tf.keras.layers.Dropout(dropout, name='Dropout_first'))
    model.add(tf.keras.layers.Dense(neurons_second, activation='relu', kernel_initializer=glorot_uniform(),
                    name='Dense_second'))
    model.add(tf.keras.layers.Dropout(dropout, name='Dropout_second'))
    model.add(tf.keras.layers.Dense(NUM_CLASSES, activation='sigmoid',
                    kernel_initializer=glorot_uniform(),
                    name='Result'))
    return model


model = getModel()
```
