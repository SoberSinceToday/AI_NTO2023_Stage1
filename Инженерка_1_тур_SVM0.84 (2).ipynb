```python
import pandas as pd
from glob import glob
```

```python
df = pd.read_csv('train.csv')
df
```

```python
arr = []
for index, row in df.iterrows():
    csv_filename = row['имя файла данных']
    arr.append(pd.read_csv(f'data/{csv_filename}').mean().rename(index))
dopdata = pd.DataFrame(arr)
df = pd.concat([df, dopdata], axis=1)
df
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
```

```python
ss = StandardScaler()
labelencoder = LabelEncoder()
df['пол'] = df['пол'].replace({'женский' : 0, 'мужской' : 1})
df['пациент off/on'] = df['пациент off/on'].replace({'off' : 0, 'on' : 1})
X_train_scaled = df.drop(['имя файла данных','Folder Path','WRIST.z'], axis=1)
y_train = np.array(df['Folder Path'])
```

```python
rf = SVC(C=1000000000, gamma=1, kernel='linear')
rf.fit(X_train_scaled, y_train)
```

```python
testData = pd.read_csv('test.csv')
arr = []
for index, row in testData.iterrows():
    csv_filename = row['имя файла данных']
    arr.append(pd.read_csv(f'data/{csv_filename}').mean().rename(index))
dopdata = pd.DataFrame(arr)
testData = pd.concat([testData, dopdata], axis=1)
testData
file_names = testData['имя файла данных']
testData['пол'] = testData['пол'].replace({'женский' : 0, 'мужской' : 1})
testData['пациент off/on'] = testData['пациент off/on'].replace({'off' : 0, 'on' : 1})
testData = testData.drop('WRIST.z', axis=1)
testData
```

```python
predict = rf.predict(testData.drop('имя файла данных', axis=1))
result = pd.DataFrame({
    'path' : file_names,
    'pred' : predict
})
result.head
```

```python
result.to_csv('submit.csv', index=False)
```
